```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
 df_can = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skipfooter=2
                      )

print('Data downloaded and read into a dataframe!')
```

```python
df_can.head()
```

```python
df_can.drop(columns=['Type','Coverage','AREA','REG','DEV','DevName'], inplace=True)
```

```python
df_can.rename(columns={'OdName':'Pais','AreaName':'Continente','RegName':'Região'}, inplace=True)
```

```python
df_can.set_index('Pais',inplace=True)
```

```python
df_can['Total'] = df_can.sum(axis=1)
```

```python
df_continent = df_can.groupby('Continente',axis=0).sum()
df_continent.head()
```

```python
df_continent['Total'].plot(kind='pie')
plt.show()
```
