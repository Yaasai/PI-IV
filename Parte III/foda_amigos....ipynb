```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import random
```

```python
# São Paulo
df_sp_3 = pd.read_excel('Sp/tab4_20_1_3.xlsx')
df_sp_3['Rede'] = 'Pública'
df_sp_3['UF'] = 'SP'
df_sp_4 = pd.read_excel('Sp/tab4_20_1_4.xls')
df_sp_4['Rede'] = 'Privada'
df_sp_4['UF'] = 'SP'

# Santa Catarina
df_sc_3 = pd.read_excel('Santa Catarina/tab4_22_1_3.xls')
df_sc_3['Rede'] = 'Pública'
df_sc_3['UF'] = 'SC'
df_sc_4 = pd.read_excel('Santa Catarina/tab4_22_1_4.xls')
df_sc_4['Rede'] = 'Privada'
df_sc_4['UF'] = 'SC'

# Minas Gerais
df_mg_3 = pd.read_excel('Minas/tab4_17_1_3.xls')
df_mg_3['Rede'] = 'Pública'
df_mg_3['UF'] = 'MG'
df_mg_4 = pd.read_excel('Minas/tab4_17_1_4.xls')
df_mg_4['Rede'] = 'Privada'
df_mg_4['UF'] = 'MG'

# Bahia
df_bh_3 = pd.read_excel('Bahia/tab4_16_1_3.xls')
df_bh_3['Rede'] = 'Pública'
df_bh_3['UF'] = 'BH'
df_bh_4 = pd.read_excel('Bahia/tab4_16_1_4.xls')
df_bh_4['Rede'] = 'Privada'
df_bh_4['UF'] = 'BH'

# Amazonas
df_am_3 = pd.read_excel('Amazonas/tab4_3_1_3.xls')
df_am_3['Rede'] = 'Pública'
df_am_3['UF'] = 'AM'
df_am_4 = pd.read_excel('Amazonas/tab4_3_1_4.xls')
df_am_4['Rede'] = 'Privada'
df_am_4['UF'] = 'AM'
```

```python
def gerar_df_todas():
    tabelas = [[df_sp_3,df_sp_4], [df_sc_3,df_sc_4], [df_mg_3,df_mg_4], [df_bh_3,df_bh_4] , [df_am_3,df_am_4]]
    tabela_final =  pd.DataFrame([], columns=df_sp_3.columns)

    for tabela in tabelas:
        for i in range(0,3):
            numero_linha = random.randint(0,tabela[0].shape[0])
            
            dados = pegar_dados(tabela[0],numero_linha)
            linha_tabela_3 = pd.DataFrame(dados, index=[numero_linha])
            tabela_final = tabela_final.append(linha_tabela_3, ignore_index=True)
            
            dados = pegar_dados(tabela[1],numero_linha)
            linha_tabela_4 = pd.DataFrame(dados, index=[numero_linha])
            tabela_final = tabela_final.append(linha_tabela_4, ignore_index=True)
    tabela_final = tabela_final.replace('-',0) 
    return tabela_final
```

```python
def pegar_dados(tabela, numero_linha):
    dados = {}
    linha = tabela.iloc[numero_linha,:]
    colunas = tabela.columns.values
    for coluna in colunas:
        dados[coluna] = linha[coluna]
        
    return dados
```

```python
final = gerar_df_todas()
final.rename(columns={final.columns[8]:'Superior'}, inplace=True)
final
```

```python
def pegar_dados_linhas(linha1, linha2):
    colunas = ['Fundamental', 'Médio', 'Superior']
    dados = []
    for coluna in colunas:
        interno = [linha1[coluna], linha2[coluna]]
        dados.append(interno)
    return dados

def pegar_dados_linha(linha1):
    colunas = ['Fundamental', 'Médio', 'Superior']
    dados = []
    for coluna in colunas:
        dados.append(linha1[coluna])
    return dados
```

```python
fig, ax = plt.subplots(5,3,figsize=[20,20])
size = 0.5
vals = np.array([[60., 32.], [37., 40.], [29., 10.]])

cmap = plt.get_cmap("tab20c")
outer_colors = cmap(np.arange(3)*4)
inner_colors = cmap(np.array([1, 2, 5, 6, 9, 10]))

k = 0
for i in range(0,5):
    for j in range(0,3):
        linha_publica = title=final.iloc[k,:]
        linha_privada = title=final.iloc[k+1,:]
        
        dados_linhas= np.array(pegar_dados_linhas(linha_publica, linha_privada))
        p1 = ax[i,j].pie(dados_linhas.sum(axis=1), radius=1, colors=outer_colors,
                   autopct='%.0f%%', wedgeprops=dict(width=size, edgecolor='w'), labels=['Fundamental', 'Médio', 'Superior'])

        p2 = ax[i,j].pie(dados_linhas.flatten(), radius=1-size, colors=inner_colors,
                     wedgeprops=dict(width=size, edgecolor='w'))
        
        ax[i,j].legend(p2[0], ('Público','Privado','Público','Privado','Público','Privado'), loc=[1,.5])
        ax[i,j].set(aspect="equal", title=final.iloc[k,11] + ' - ' + final.iloc[k,10])
        k += 2

plt.savefig('foo.png')
plt.show()
```

```python
fig, axs = plt.subplots(5,3,figsize=[20,20])
width = 0.5
k = 0
for i in range(0,5):
    for j in range(0,3):
        
        linha_publica = title=final.iloc[k,:]
        linha_privada = title=final.iloc[k+1,:]
        dados_publica = pegar_dados_linha(linha_publica)
        dados_privada = pegar_dados_linha(linha_privada)
        
        p1 = axs[i,j].bar(np.arange(3), dados_publica, width)
        axs[i,j].bar(np.arange(3), dados_privada, width, bottom=dados_publica)
        axs[i,j].set_xticks(np.arange(3))
        axs[i,j].set_xticklabels(['Fundamental', 'Médio', 'Superior'])
        axs[i,j].set_title(final.iloc[k,11] + ' - ' + final.iloc[k,10])
        axs[i,j].legend((p1[0], p2[0]), ('Público', 'Privado'))
        k += 2
        
plt.savefig('foo.png')
plt.show()
```
