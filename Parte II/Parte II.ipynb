```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Lendo a tabela
df = pd.read_csv('./Tabelas PI IV/Zonal annual means, 1880-present.csv')
df.head()
```

```python
# Criando uma tabela que vai pegar somente um subconjunto dos dados
df_global = df.iloc[:,0:4].copy()
df_global.rename(columns={'Glob':'Global', 'NHem':'Hemisfério norte', 'SHem':'Hemisfério sul'}, inplace=True)
df_global.set_index('Year', inplace=True)
df_global.head()
```

```python
ax = df_global.plot(kind='box', figsize=[10,10], color='#292626')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

#plt.axhline(0, c="#d4d2d2",lw=0.9,zorder=1)
plt.ylabel('Alteração da temperatura média anual da terra')
plt.xlabel('Ano')
plt.savefig('./Relatorio/boxplot.png')
plt.show()
```

```python
ax = df_global.plot(kind='line', figsize=[10,10], color=['#5cb85c','#5bc0de','#d9534f'],lw=3.3)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.axhline(0, c="#d4d2d2",lw=0.9,zorder=1)
plt.ylabel('Alteração da temperatura média anual da terra')
plt.xlabel('Ano')
# SAVE PICTURE
plt.savefig('./Relatorio/lines.png')
plt.show()
```

```python
df_scatter = df.iloc[:,0:4].copy()
df_scatter.rename(columns={'Glob':'Global', 'NHem':'Hemisfério norte', 'SHem':'Hemisfério sul'}, inplace=True)

fig = plt.figure(figsize=[10,10])

ax = fig.add_subplot(111)

year = df_scatter['Year']
mudancaGlobal = df_scatter['Global']
hemisferioS = df_scatter['Hemisfério sul']
hemisferioN = df_scatter['Hemisfério norte']


ax.scatter(y=year, x=mudancaGlobal, lw=3.3, label='Global')
ax.scatter(y=year, x=hemisferioS, lw=3.3, label='Hemisfério sul')
ax.scatter(y=year, x=hemisferioN, lw=3.3, label='Hemisfério norte')


ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.ylabel('Ano')
plt.xlabel('Alteração da temperatura média anual da terra')
plt.savefig('./Relatorio/dots.png')
plt.show()
```
